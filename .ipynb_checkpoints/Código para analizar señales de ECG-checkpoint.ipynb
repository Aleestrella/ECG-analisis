```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from scipy.io import loadmat
```

```python
x=loadmat('100m.mat')
```

```python
ecg1=(x['val']-0)/200
```

```python
ecg=np.transpose(ecg1)
```

```python
fs=360
```

```python
tm=1/fs
```

```python
t=np.linspace(0,np.size(ecg),np.size(ecg))*tm
```

```python
plt.plot(t, ecg)
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud (mV)')
plt.title('ECG')
plt.grid(True)
plt.show()
```
